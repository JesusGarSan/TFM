```python
# Importación de librerías
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

## FIG. S1

```python
N = 3 # Número de agentes
x = np.ones(N) * 1000
chi = np.zeros(N)
a = np.ones(N) * 0.5
a_defection = np.zeros(N)
sigma = .2
mu = 1.01

# Todos los agentes cooperan igual excepto el 1º, que es el que vamos cambiando
a[0] = 0.1

```

```python
chi = np.random.normal(1, 0.1, 100000)

print(abs(1 - np.mean(chi)))
print(abs(sigma - np.std(chi, ddof=1)))
```

```python
steps = 10000
time = range(steps)
X = np.zeros((steps, N))
X
rel_growth_rate = np.zeros((steps, N))

x_old = x
x_defection_old = x
for i in time:
    chi = np.random.normal(mu, sigma, N)
    x = x_old*chi *(1-a) + np.mean(a*x_old*chi)
    x_defection = x_defection_old*chi 
    X[i] = x
    rel_growth_rate[i] = (x-x_defection)/x

    x_old = x
    x_defection_old = x
```

```python
plt.plot(X)

rel_growth_rate[-1]
```
